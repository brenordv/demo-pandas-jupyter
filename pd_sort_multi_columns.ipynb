# Loading Sample Data

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/brenordv/data-samples/main/csv/NASA/missionstars.csv", comment="#")
```

```python
df.head()
```

# Sorting multiple columns - all ASCENDING

```python
df.sort_values(["star_name", "hd_name", "tm_name"])
```

```python
df.sort_values(["star_name", "hd_name", "tm_name"], ascending=True)  # explicitly setting ascending to True, which is the default value.
```

# Sorting multiple columns - all DESCENDING

```python
df.sort_values(["star_name", "hd_name", "tm_name"], ascending=False)
```

# Sorting multiple columns - Some Ascending, some descending

```python
df.sort_values(
    ["star_name", "hd_name", "tm_name"],
    ascending=[
        False,  # star_name
        True,   # hd_name
        False   # tm_name
    ])
```
